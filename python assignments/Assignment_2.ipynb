8. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.

```python

try:
    spam = int(input("Enter an integer\n"))
    
    if (spam==1):
        print("Hello")
    elif (spam==2):
        print("Howdy")
    else:
        print("Greetings")
except Exception as e:
    print("please provide a valid integer value")
    print(e)
```

10. How can you tell the difference between break and continue?

Sol )  
break statement helps to break out of the loop. continue statement continues to the next iteration  
For example, the below code can be used to filter odd numbers only  

```python
try:
    num = int(input("Enter an integer\n"))
    odd_lst = []
    
    for i in range(num):
        if (i%2==0):
            continue  # continue to the next iteration if i is even
        odd_lst.append(i)
    print(odd_lst)
except Exception as e:
    print("please provide a valid integer value")
    print(e)
```

In the below code, since we have written break statement, it will break out of the loop on the first occurrence of when i is even 

```python
try:
    num = int(input("Enter an integer\n"))
    odd_lst = []
    
    for i in range(1, num):
        if (i%2==0):
            break  # break from the loop on first encounter of an even num
        odd_lst.append(i)
    print(odd_lst)
except Exception as e:
    print("please provide a valid integer value")
    print(e)
```

11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Sol )  
range(10) is synonymous to range(0, 10) and range(0, 10, 1)  
Since by default, the initial value and step size are 0 and 1 respectively  
In range(10) , we have not specified the initial value ( therefore default to 0), and not specified step size ( therefore default to 1)  
In range(0, 10) , we have specified the initial value but not specified step size ( therefore, default to 1)  
In range(0, 10, 1) , we have specified the initial value as well as the step size  

12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

```python
# Sol ) 
print("Printing 1 to 10 with for loop")

for i in range(1, 11):
    print(i)

print("Printing 1 to 10 with while loop")
num = 1
while(num<=10):
    print(num)
    num += 1
```
